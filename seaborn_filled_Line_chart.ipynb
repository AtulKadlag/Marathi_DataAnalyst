```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Flights dataset
flights = sns.load_dataset('flights')

# Pivot the dataset for better plotting
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

# Setting the plot size for better visibility
plt.figure(figsize=(10, 6))

# Looping through each year to create individual lines
for year in flights_pivot.columns:
    plt.plot(flights_pivot.index, flights_pivot[year], label=year)

# Filling under the line plot
plt.fill_between(flights_pivot.index, flights_pivot.min(axis=1), flights_pivot.max(axis=1), alpha=0.3)

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.title('Monthly Total Passengers (1949-1960)')
plt.legend(title='Year')

# Display the plot
plt.show()
```
