```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading data from CSV
data = pd.read_csv('sales_data_1.csv')

# Creating the line chart with customizations
plt.figure(figsize=(10, 6))  # Set figure size

plt.plot(data['Month'], data['Sales'], marker='^', linestyle='--', color='b', label='Sales Data')  # Customized line and marker

# Adding grid, title, labels, and legend
plt.grid(True)  # Add grid
plt.title('Monthly Sales Data', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.legend()  # Add legend

# Display the plot
plt.tight_layout()  # Adjust layout to not cut off labels
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Reading data from CSV
data = pd.read_csv('sales_data_1.csv')

# Creating the line chart with customizations
plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(data['Month'], data['Sales'], marker='^', linestyle='--', color='b', label='Sales Data')  # Customized line and marker

# Annotating each data point with the sales value
for i, txt in enumerate(data['Sales']):
    plt.annotate(txt, (data['Month'][i], data['Sales'][i]), textcoords="offset points", xytext=(0,10), ha='center')

# Adding grid, title, labels, and legend
plt.grid(True)  # Add grid
plt.title('Monthly Sales Data', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.legend()  # Add legend

# Display the plot
plt.tight_layout()  # Adjust layout to not cut off labels
plt.show()
```
