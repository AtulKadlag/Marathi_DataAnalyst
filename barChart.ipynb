```python
import matplotlib.pyplot as plt

# Data
products = ['Mobiles', 'Accessories', 'Laptops']
sales = [23, 38, 12]


# Creating the bar chart
plt.bar(products, sales)

# Adding title and labels
plt.title('Sales Data')
plt.xlabel('Products')
plt.ylabel('Sales')

# Display the plot
plt.show()

```
